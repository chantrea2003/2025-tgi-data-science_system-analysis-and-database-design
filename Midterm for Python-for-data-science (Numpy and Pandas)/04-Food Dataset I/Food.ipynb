<b>Exercise 5. Display the <i>"food"</i> which has the highest number of orders?

```python
print("The most ordered food item is:", df['food'].value_counts().idxmax())
```

<b>Exercise 5. How many items were ordered for the most-ordered item?

```python
most_order = df['food'].value_counts().max()
print("The highest number of orders for a single food item is:", most_order)
```

<b>Exercise 6. Display the <i>"description"</i> which has the highest number of orders?</b>

```python
print("The most ordered food item is:", df['description'].value_counts().idxmax())

```

<b>Exercise 7. How many foods ordered in total?</b>

```python
Ordered_in_total = df['quantity'].sum()
print("The total quantity of food items ordered is:", Ordered_in_total)
```

<b>Exercise 8. Calculate the total income for the period in the dataset?</b>

```python
df['unit_price'] = df['unit_price'].str.replace('$', '', regex=False).astype(float)

total_income = (df['quantity'] * df['unit_price']).sum()
print("The total income from food items ordered is:", total_income,"$")
```

<b>Exercise 9. Find the total number of orders.</b>

```python
print("The total number of food orders is:", len(df))
```

<b>Exercise 10. Find the average income for each order?</b>

```python
avg = avg = total_income / len(df)
print("The average income per order is:", avg.round(2), "$") 
```
