# using looop


```python
numbers = [100,2,3,4,5]

min = numbers[0]
max = numbers[0]

for num in numbers:
    if num < min:
        min = num
    if num > max:
        max = num
    total = sum(numbers)
    average = total / len(numbers)

print("Min: ", min, " Max: ", max, " sum: ", total, " Average: ", average)

```

Print shapes

```python
rows = 3
cols = 5

i = 0
while i < rows:
    j = 0
    while j < cols:
        print("*", end=" ")
        j += 1
    print()
    i += 1
```

```python
height = 5

i = 1
while i <= height:
    j = 1
    while j <= i:
        print("*", end=" ")
        j += 1
    print()
    i += 1
```

Write a few functions to:
Add a new movie to your collection (consider using appropriate data structures).
Search for movies by title or director.
Filter movies based on genre or year of release.
Display movie information in a user-friendly format


```python
movies = [
    {
        "title": "The Shawshank Redemption",
        "director": "Frank Darabont",
        "year": 1994,
        "genre": ["Drama"]
    },
    {
        "title": "The Godfather",
        "director": "Francis Ford Coppola",
        "year": 1972,
        "genre": ["Crime", "Drama"]
    }
]
```

Add movie

```python
def add_movie(collection, title, director, year, genre):
    new_movie = {
        "title": title,
        "director": director,
        "year": year,
        "genre": genre
    }
    collection.append(new_movie)
```

```python
add_movie(movies, "Inception", "Christopher Nolan", 2025, ["Drama", "Action"])
```

Search movie

```python
def search_movies(collection, keyword):
    keyword = keyword.lower()
    result = []

    for m in collection:
        if keyword in m["title"].lower() or keyword in m["director"].lower():
            result.append(m)

    return result
```

```python
search_movies(movies, "godfather")
```

```python
search = lambda collection, k: list(
    filter(lambda m: k.lower() in m["title"].lower() 
                    or k.lower() in m["director"].lower(), collection)
)
search(movies, "nolan")

```

Filter by genre or year

```python
def filter_by_genre(collection, genre):
    return list(filter(lambda m: genre in m["genre"], collection))

def filter_by_year(collection, year):
    return list(filter(lambda m: m["year"] == year, collection))
```

```python
filter_by_year(movies, 1994)
```

Sort by year

```python
def sort_by_year(collection):
    return sorted(collection, key=lambda m: m["year"])
```

```python
sort_by_year(movies)
```

Display movie

```python
def display(movie):
    print(f"{movie['title']} ({movie['year']})")
    print(f"Director: {movie['director']}")
    print(f"Genre: {', '.join(movie['genre'])}")

```
